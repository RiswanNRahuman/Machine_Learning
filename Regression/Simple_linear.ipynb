```python
import numpy as numpy
import pandas as pd
import matplotlib.pyplot as plt
area=[2600,2800,3100,3400,3800]
price=[5600000,5800000,6100000,6600000,7200000]
```

```python
plt.scatter(area,price,color='c')
plt.xlabel('Area',color='r',size=13)
plt.ylabel('Price',color='r',size=13)
plt.title('Area-Price graph',color='r',size=15)
```
