```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('company_sales_data.csv')
df
```

```python
x1=df['month_number']
y1=df['total_profit']
plt.plot(x1,y1)
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.xticks(x1)
plt.yticks([100000,200000,300000,400000,500000])
```

```python
plt.plot(x1,y1,'o:r',linewidth=3,label="Profit Data")
plt.xlabel('Month Number')
plt.ylabel('Solid Unit Number')
plt.legend()
plt.xticks(x1)
plt.yticks([100000,200000,300000,400000,500000])
```

```python
fcream=df["facecream"]
fwash=df["facewash"]
tpaste=df["toothpaste"]
bsoap=df["bathingsoap"]
shamp=df["shampoo"]
moist=df["moisturizer"]
plt.plot(x1,cream,marker='s',color='c')
```

```python
products=["facecream","facewash","toothpaste","bathingsoap","shampoo","moisturizer"]
for i in products:
    plt.plot(df["month_number"],df[i],marker="s",color="g")
plt.xlabel("Month Number")
plt.ylabel("Product sales")
plt.title("Multline Plot for all product sales")

```

```python
plt.scatter(df["month_number"],df["toothpaste"])
plt.xlabel("Month Number")
plt.ylabel("Toothpaste Sales")
plt.title("Toothpaste Sales data of each month")

```

```python
plt.bar([a-0.25 for a in x1],df["facecream"],width=0.25,align='edge',label="Face Cream")
plt.bar([a+0.25 for a in x1],df["facewash"],width=-0.25,align='edge',label="Face Wash")
plt.grid()
plt.xlabel("Month number")
plt.ylabel("Facecream and Facewash Sales")
plt.xticks(x1)
plt.legend(loc='upper right')
```

```python
plt.hist(df["total_profit"])
plt.xlabel("Profit Range")
plt.ylabel("Frequency")
plt.title("Histogram of Total Profits")
```

```python
total_sales=df[products].sum()
plt.pie(total_sales,autopct="%1.01f%%",labels=products)
plt.legend(loc="upper left")
plt.tight_layout()
plt.title("Total Sales data for last year for each product")
```
