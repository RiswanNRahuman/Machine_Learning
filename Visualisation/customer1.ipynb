```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('customer1.txt',header=None)
df
```

```python
df.columns=['id','Fname','Lname','Age','Proffession','Location']
df
```

```python
df1=df['Location'].value_counts()
df1
```

```python
plt.bar(df1.index,df1.values,color='c')
plt.xlabel('Location',color='r')
plt.ylabel('Count',color='r')
plt.title('location-count data',color='k')
plt.grid()
```

```python
labs=df1.index
plt.pie(df1.values,labels=labs,autopct='%1.0f%%')
plt.legend()
plt.tight_layout()
```
